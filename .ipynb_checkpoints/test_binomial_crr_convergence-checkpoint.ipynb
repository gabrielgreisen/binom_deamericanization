```python
# imports

import numpy as np
import matplotlib.pyplot as plt
import binom_crr
```

```python
european_call = []

for n in range(10,1000,10):
    european_call.append(binom_crr.binomial_crr("C", 100, 1, 100, 0.2, 0.05, 0.02, n, "E"))
```

```python
plt.plot(range(10,1000,10), np.array(european_call), label='european call')
plt.xlabel("Time Step (N)")
plt.ylabel("Predicted options pricing")
plt.axhline(y=9.22701, color='r', ls='--', label='exact price @ 9.22701')
plt.legend()
plt.title("Convergence of european call options pricing over time step")
```

```python
american_put = []
for n in range(10,10000,100):
    american_put.append(binom_crr.binomial_crr("P", 100, 1, 100, 0.2, 0.05, 0.02, n, "A"))
```

```python
plt.plot(range(10,10000,100), np.array(american_put), label = "american put")
plt.axhline(y=6.66069, color='r', ls="--", label="exact price @ 6.66069")
plt.legend()
plt.title("Convergence of american put options pricing over time step")
```
